# 6. Identify the significant variables by performing statistical analysis using p-values

```python
from scipy import stats
```

```python
df_test = sr[['Complaint Type', 'Time_Elapsed_seconds']].dropna()
```

```python
groups = [
    group['Time_Elapsed_seconds'].values
    for name, group in df_test.groupby('Complaint Type')
]
```

```python
f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
# Complaint Type is significant (affects response time) because p value <0.5
```

```python
# “I performed ANOVA to test whether the mean response time differs across complaint types. Since the p-value was less than 0.05, I concluded that complaint type significantly impacts response time.
```

# 7. Perform a Kruskal-WallisH test
## 7.1Fail to reject H0: All sample distributions are equal
## 7.2Reject H0: One or more sample distributions are not equal

```python
df_test = sr[['Complaint Type', 'Time_Elapsed_seconds']].dropna()
```

```python
top5 = sr['Complaint Type'].value_counts().head(5).index
df_test = df_test[df_test['Complaint Type'].isin(top5)]
```

```python
groups = [
    group['Time_Elapsed_seconds'].values
    for name, group in df_test.groupby('Complaint Type')
]
```

```python
h_stat, p_value = stats.kruskal(*groups)

print("H-statistic:", h_stat)
print("P-value:", p_value)
```

```python
p_value < 0.05
```

```python
#Reject H0: One or more sample distributions are not equal”
```

```python
# # Since the data was not normally distributed, I used the Kruskal-Wallis test. Based on the p-value, I determined whether 
# complaint types have significantly different response time distributions
```
